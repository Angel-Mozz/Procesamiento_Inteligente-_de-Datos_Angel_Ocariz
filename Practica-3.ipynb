4. Estimar el corto de calefacción para una Temperatura media de 30°, 5 pulgadas de aislamiento y un calentador con antiguedad de 10 años

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv("SalsberryRealty.csv")
df.head()
```

### 1. Determinar las variables independientes
x1 = temp, x2 = Insul, x3 = Age

```python
X = df[['Temp', 'Insul', 'Age']]
```

### 2. Determinar las variables dependientes Y
y = cost

```python
Y = df[['Cost']]
```

```python
#Ajustar el modelo de regresion multiple
model = LinearRegression()
model.fit(X, Y) #Entrenando el modelo
y_pred = model.predict(X)
y_pred

#Metricas estadisticas
mse = mean_squared_error(Y, y_pred) #Error de la media caudratica
r2 = r2_score(Y, y_pred) #Varianza
```

### 3. Valor de interseccion

```python
model.intercept_
```

```python
#Coeficientes
print('Coeficientes: ', model.coef_)
print('Error cuadratico medio: ', mse)
print('Varianza: ', r2)
```

### 4. Estimar para 30 °, aislamiento 5", antiguedad 10 años

```python
Y = model.predict([[30, 5, 10]])
print("El costo de calefaccion estimado es: ", Y)
```
