```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_players = 100000
precio = np.random.choice(['Alto', 'Medio', 'Bajo'], size = n_players, p=[0.20, 0.30, 0.50])
genero_juego = np.random.choice(['Accion', 'Aventura', 'Estrategia', 'Otros'], size = n_players, p=[0.40, 0.30, 0.20, 0.10])
amigos = np.random.choice(['Si', 'No'], size = n_players, p=[0.65, 0.35])
compra = np.zeros(n_players)
```

```python
for i in range(n_players):

    # PRECIO Alto
    if precio[i] == 'Alto' and amigos[i] == 'Sí' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.06, 0.94])
    elif precio[i] == 'Alto' and amigos[i] == 'Sí' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.05, 0.95])
    elif precio[i] == 'Alto' and amigos[i] == 'Sí' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.04, 0.96])
    elif precio[i] == 'Alto' and amigos[i] == 'Sí' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.03, 0.97])

    elif precio[i] == 'Alto' and amigos[i] == 'No' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.03, 0.97])
    elif precio[i] == 'Alto' and amigos[i] == 'No' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.02, 0.98])
    elif precio[i] == 'Alto' and amigos[i] == 'No' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.015, 0.985])
    elif precio[i] == 'Alto' and amigos[i] == 'No' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.01, 0.99])

    # PRECIO MEDIO
    elif precio[i] == 'Medio' and amigos[i] == 'Sí' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.15, 0.85])
    elif precio[i] == 'Medio' and amigos[i] == 'Sí' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.12, 0.88])
    elif precio[i] == 'Medio' and amigos[i] == 'Sí' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.10, 0.90])
    elif precio[i] == 'Medio' and amigos[i] == 'Sí' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.08, 0.92])

    elif precio[i] == 'Medio' and amigos[i] == 'No' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.08, 0.92])
    elif precio[i] == 'Medio' and amigos[i] == 'No' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.06, 0.94])
    elif precio[i] == 'Medio' and amigos[i] == 'No' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.05, 0.95])
    elif precio[i] == 'Medio' and amigos[i] == 'No' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.03, 0.97])

    # PRECIO BAJO
    elif precio[i] == 'Bajo' and amigos[i] == 'Sí' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.25, 0.75])
    elif precio[i] == 'Bajo' and amigos[i] == 'Sí' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.22, 0.78])
    elif precio[i] == 'Bajo' and amigos[i] == 'Sí' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.20, 0.80])
    elif precio[i] == 'Bajo' and amigos[i] == 'Sí' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.18, 0.82])

    elif precio[i] == 'Bajo' and amigos[i] == 'No' and genero_juego[i] == 'Acción':
        compra[i] = np.random.choice([1, 0], p=[0.12, 0.88])
    elif precio[i] == 'Bajo' and amigos[i] == 'No' and genero_juego[i] == 'Aventura':
        compra[i] = np.random.choice([1, 0], p=[0.10, 0.90])
    elif precio[i] == 'Bajo' and amigos[i] == 'No' and genero_juego[i] == 'Estrategia':
        compra[i] = np.random.choice([1, 0], p=[0.08, 0.92])
    elif precio[i] == 'Bajo' and amigos[i] == 'No' and genero_juego[i] == 'Otros':
        compra[i] = np.random.choice([1, 0], p=[0.05, 0.95])

    else:
        compra[i] = np.random.choice([1, 0], p=[0.001, 0.999])
```

```python
data = pd.DataFrame({'Precio': precio, 'Amigos': amigos, 'Genero_Juego': genero_juego, 'Compra': compra})

probabilidad_compra = data.groupby(["Precio", "Amigos", "Genero_Juego"])["Compra"].mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(14, 12))

probabilidad_compra["Etiqueta"] = (
    probabilidad_compra["Precio"].astype(str)
    + " , "
    + probabilidad_compra["Amigos"].astype(str)
    + " , "
    + probabilidad_compra["Genero_Juego"].astype(str)
)
probabilidad_compra.plot(kind="barh", x="Etiqueta", y="Compra", ax=ax, legend=True)

for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f"{v:.5f}" for v in container.datavalues],
        label_type="edge",
        padding=3
    )

ax.set_xlabel("P(Compra = Sí)")
ax.set_title("Probabilidades Condicionales de Compra")

plt.tight_layout()
#plt.savefig("probabilidades_condicionales_compra.png", dpi=300)
plt.show()
```
